# CODE_pour_grand_lien_page principale

```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Créez une instance du pilote WebDriver Chrome
driver = webdriver.Chrome()

# Initialisez une liste vide pour stocker les données
annonces_data = []
num_resultat = 0  # Compteur pour numéroter les résultats

for page_num in range(1, 498):  # Boucle de la page 1 à la page 497
    # Construisez l'URL de la page en fonction du numéro de page
    url = f"https://deals.jumia.sn/appartements-a-vendre?xhr=4ibtg&page={page_num}"
    
    # Accédez à la page web
    driver.get(url)

    # Attendre un peu pour que la page se charge complètement (ajuster le temps d'attente si nécessaire)
    time.sleep(5)

    # Récupérez toutes les annonces sur la page actuelle
    annonces = driver.find_elements(By.XPATH, '//*[@id="search-results"]/article')

    for annonce in annonces:
        num_resultat += 1  # Incrémentez le compteur
        titre_element = annonce.find_element(By.XPATH, './/div[2]/div/div[1]/a/span')
        price_element = annonce.find_element(By.XPATH, './/div[2]/div/div[2]/span')
        lieu_element = annonce.find_element(By.XPATH, './/div[2]/div/div[1]/span')
        date_element = annonce.find_element(By.XPATH, './/div[2]/div/div[2]/time')

        # Obtenez le texte des éléments ou utilisez "NA" s'ils ne sont pas disponibles
        titre = titre_element.text if titre_element else "NA"
        price = price_element.text if price_element else "NA"
        lieu = lieu_element.text if lieu_element else "NA"
        date = date_element.text if date_element else "NA"

        # Ajoutez les données de l'annonce actuelle au dictionnaire
        annonce_data = {
            "Numéro": num_resultat,
            "Type de logement": titre,
            "Prix": price,
            "Lieu": lieu,
            "Date_pub": date
        }

        annonces_data.append(annonce_data)

        print(f"Résultat {num_resultat}:")
        for key, value in annonce_data.items():
            print(f"{key}: {value}")
        print("-" * 50)

# Fermez le navigateur lorsque vous avez terminé
driver.quit()

# Définir le nom du fichier CSV
filename = "C:/Users/lenovo/Desktop/Dolph stats/dolph stats/Barometre immobilier/Barometre Immobilier/Barometer_Jeff/Jumia_appart_vendre.csv"

# Ouvrir le fichier CSV en mode écriture
with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
    # Créez un écrivain de dictionnaire CSV
    fieldnames = ["Numéro", "Type de logement", "Prix", "Lieu", "Date_pub"]
    writer = csv.DictWriter(csvfile, delimiter=';', fieldnames=fieldnames)

    # Écrivez l'en-tête du fichier CSV
    writer.writeheader()

    # Écrivez les données dans le fichier CSV
    writer.writerows(annonces_data)

print(f"Les données ont été stockées dans le fichier CSV : {filename}")
```
