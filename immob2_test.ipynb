## Scrapping de tous les sous liens des appartement sur le site

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Créez une instance du pilote WebDriver Chrome
driver = webdriver.Chrome()

# Liste pour stocker les données des annonces
annonces_data = []

# Fonction pour extraire les informations d'une page de détail
def extract_info_detail(url):
    driver.get(url)
    time.sleep(5)
    
    titre = driver.find_element(By.XPATH, '//*[@id="main-holder"]/article/header/h1/span').text
    price = driver.find_element(By.XPATH, '//*[@id="priceSection"]/span').text
    vendeur = driver.find_element(By.XPATH, '//*[@id="priceSection"]/div/dl/dd[1]/span').text
    lieu = driver.find_element(By.XPATH, '//*[@id="priceSection"]/div/dl/dd[2]/span').text
    date = driver.find_element(By.XPATH, '//*[@id="priceSection"]/div/dl/dd[3]/time').text
    nombre_de_pieces = driver.find_element(By.XPATH, '//*[@id="main-holder"]/article/div[1]/div/div[2]/div/h3[1]/span').text
    superficies = driver.find_element(By.XPATH, '//*[@id="main-holder"]/article/div[1]/div/div[2]/div/h3[2]/span').text
    unite_de_superficie = driver.find_element(By.XPATH, '//*[@id="main-holder"]/article/div[1]/div/div[2]/div/h3[2]/span/sup').text
    description = driver.find_element(By.XPATH, '//*[@id="main-holder"]/article/div[1]/div/div[3]/p').text
    tel_vendeur = driver.find_element(By.XPATH, '//*[@id="contactPhone"]/div/div[1]/a').text
    
    annonce_data = {
        "Titre": titre,
        "Prix": price,
        "Vendeur": vendeur,
        "Lieu": lieu,
        "Date": date,
        "Nombre de pièces": nombre_de_pieces,
        "Superficie": superficies,
        "Unité de superficie": unite_de_superficie,
        "Description": description,
        "Téléphone du vendeur": tel_vendeur
    }
    
    annonces_data.append(annonce_data)

# Parcourir les pages de la page principale de 1 à 2
for page_num in range(1, 3):
    main_page_url = f"https://deals.jumia.sn/appartements-a-vendre?page={page_num}"
    driver.get(main_page_url)
    time.sleep(5)
    
    # Récupérer les liens des annonces sur la page principale
    annonce_elements = driver.find_elements(By.XPATH, '//*[@id="search-results"]/article/div/div[2]/div/div[1]/a')
    
    for i, annonce_element in enumerate(annonce_elements):
        annonce_url = annonce_element.get_attribute("href")
        extract_info_detail(annonce_url)
        
        # Afficher les informations extraites pour cette annonce en temps réel
        print("-" * 50)
        print(f"Annonce {i + 1} (Page {page_num}):")
        for key, value in annonces_data[-1].items():
            print(f"{key}: {value}")
        print("-" * 50)

# Fermer le navigateur lorsque vous avez terminé
driver.quit()
```
